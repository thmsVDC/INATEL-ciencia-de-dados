```python
import pandas as pd
import numpy as np
from IPython.display import display
import matplotlib.pyplot as plt

matches = pd.read_csv("WorldCupMatches.csv")
cups = pd.read_csv("WorldCups.csv")
```

## Quais países-sede geraram maior audiência média por partida ao longo das Copas do Mundo?
- Ajuda a FIFA a escolher sedes que historicamente atraem mais público (TV, online, in loco).
- Dá argumentos para patrocinadores investirem mais quando a Copa acontece nesses países, já que o retorno em exposição tende a ser maior.


```python
import pandas as pd
import matplotlib.pyplot as plt

grouped_cups = cups.groupby('Country', as_index=False)[['Attendance', 'MatchesPlayed']].sum()
grouped_cups['Attendance'] = grouped_cups['Attendance'].astype(float)
grouped_cups['AvgAttendancePerMatch'] = grouped_cups['Attendance'] / grouped_cups['MatchesPlayed']
grouped_cups_sorted = grouped_cups.sort_values(by='AvgAttendancePerMatch', ascending=False)

fig, ax = plt.subplots(figsize=(14, 6))
bars = ax.bar(grouped_cups_sorted['Country'], grouped_cups_sorted['AvgAttendancePerMatch'], color='seagreen')

ax.bar_label(bars, fmt='%.0f')
ax.set_title('Audiência Média por Partida por País-Sede (Copa do Mundo)', fontsize=14, fontweight='bold')
ax.set_ylabel('Público Médio por Jogo')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Por que os EUA estão em primeiro?
Os **EUA** estão em primeiro devido à **grande capacidade de público** dos estádios, à **popularidade crescente do futebol** no país e ao **impacto positivo** da Copa de 1994, que ajudou a expandir a base de fãs. Além disso, a **infraestrutura moderna** do país contribui para o alto número de espectadores.

### Por que a África do Sul está entre os primeiros?
A **África do Sul** teve uma grande **importância histórica** por ser o primeiro país africano a sediar a Copa. A **visibilidade internacional**, os **investimentos em infraestrutura** e o **legado pós-apartheid** aumentaram a audiência e o engajamento.

# Há tendência de aumento na média de público ao longo dos anos?
- Ver se o evento tem ganhado popularidade e justificar maiores investimentos.


```python
cups['Attendance'] = cups['Attendance'].astype(float)
cups['AvgAttendancePerMatch'] = cups['Attendance'] / cups['MatchesPlayed']

plt.figure(figsize=(12, 6))
plt.plot(cups['Year'], cups['AvgAttendancePerMatch'], marker='o', color='royalblue', linewidth=2)
plt.title('Tendência de Público Médio por Partida ao Longo dos Anos', fontsize=14, fontweight='bold')
plt.xlabel('Ano')
plt.ylabel('Público Médio por Jogo')
plt.xticks(cups['Year'], rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### Há tendência de aumento na média de público ao longo dos anos?

Sim, há uma tendência clara de **aumento da média de público** ao longo dos anos. Isso pode ser atribuído a vários fatores:

### 1. **Crescimento global da popularidade do futebol**:
   O futebol se tornou mais **popular** e **globalmente acessível**, atraindo um público crescente, tanto no **estádio** quanto pela **transmissão televisiva**.

### 2. **Melhora na infraestrutura**:
   Países-sede passaram a investir em **estádios modernos**, **transporte eficiente** e **acessibilidade** para garantir uma melhor **experiência ao público**, permitindo maior capacidade de público.

### 3. **Aumento do número de times e jogos**:
   A ampliação do torneio, como o aumento de **times qualificados** e o número de **partidas disputadas**, impacta diretamente o volume de público, com **mais jogos atraindo mais espectadores**.

### 4. **Exposição e mídia**:
   O **crescimento da cobertura midiática**, incluindo a **internet** e **redes sociais**, aumentou o alcance do evento, atraindo mais pessoas para os estádios e gerando **maior visibilidade global**.

### 5. **Pico de público em 1950**:
   Em 1950, o **Brasil** foi o país-sede e o evento aconteceu após a **Segunda Guerra Mundial**, quando o mundo estava se recuperando. A Copa foi um símbolo de **renascimento** e o Brasil, com uma grande paixão pelo futebol, atraiu uma enorme audiência. Além disso, a **final histórica** entre Brasil e Uruguai, no Maracanã, foi um dos maiores eventos esportivos da história, com um público recorde de mais de **200 mil pessoas**.

### 6. **Pico de público em 1994**:
   O auge de público na Copa de 1994 nos Estados Unidos pode ser explicado pela **primeira vez em que o torneio aconteceu no país**, que não tinha uma cultura futebolística forte na época. A **demanda gerada pela novidade** e o **apelo do evento internacional** atraiu uma grande audiência, junto com o aumento de **divulgação na mídia** e a **globalização do esporte**.

   ### Por que, após 1950, houve 3 edições com pouquíssima audiência?

### 7. **Fatores Políticos e Econômicos**:
   Em 1954, o cenário global ainda estava sendo marcado pela **recuperação pós-guerra** e a **polarização da Guerra Fria**. A Copa na **Suíça** (1954) não teve o mesmo apelo em comparação com o Brasil, e a **escassez de transporte e infraestrutura** dificultou o acesso de fãs internacionais.

### 8. **Localizações Menos Acessíveis**:
   Copas como a de **Suíça (1954)** e **Chile (1962)** foram realizadas em **localizações menos acessíveis** em comparação com outras edições. A **infraestrutura limitada** e as dificuldades de viagem diminuíram a capacidade de atrair grandes públicos internacionais, especialmente em um contexto de **pós-guerra** e **crises econômicas**.

### 9. **Ausência de Estádios Modernos e Mídia Menos Forte**:
   Durante essas edições, a **cobertura midiática** e a **infraestrutura de estádios** não eram tão desenvolvidas, o que restringia o alcance do evento e seu poder de atrair audiências massivas.

# Existe relação entre país-sede e desempenho do time da casa?
- Ver se o time do país-sede costuma ir longe (mais exposição nacional).



```python
ajusted_cups = cups.copy()
ajusted_cups['Country'] = ajusted_cups['Country'].replace({
    'Germany FR': 'Germany',
    'Korea/Japan': 'Korea Republic'
})

japan_row = ajusted_cups[ajusted_cups['Year'] == 2002].copy()
japan_row['Country'] = 'Japan'
ajusted_cups = pd.concat([ajusted_cups, japan_row], ignore_index=True)

hosts = ajusted_cups[['Year', 'Country']]
host_performance = []

for _, row in hosts.iterrows():
    year = row['Year']
    country = row['Country']

    year_matches = matches[matches['Year'] == year]

    host_games = year_matches[
        (year_matches['Home Team Name'] == country) |
        (year_matches['Away Team Name'] == country)
    ]

    wins, draws, losses = 0, 0, 0
    for _, match in host_games.iterrows():
        home_team = match['Home Team Name']
        away_team = match['Away Team Name']
        home_goals = match['Home Team Goals']
        away_goals = match['Away Team Goals']

        if home_team == country:
            if home_goals > away_goals:
                wins += 1
            elif home_goals == away_goals:
                draws += 1
            else:
                losses += 1
        elif away_team == country:
            if away_goals > home_goals:
                wins += 1
            elif away_goals == home_goals:
                draws += 1
            else:
                losses += 1

    host_performance.append({
        'Country': country,
        'Year': year,
        'Wins': wins,
        'Draws': draws,
        'Losses': losses
    })

df_perf = pd.DataFrame(host_performance)

df_perf_sorted = df_perf.groupby('Country').agg({'Wins': 'sum'}).sort_values(by='Wins', ascending=False)
top_10 = df_perf_sorted.head(10)

top_10.plot(
    kind='bar',
    stacked=True,
    figsize=(12, 6),
    colormap='tab20c'
)
plt.title('Top 10 Países com Melhor Desempenho como País-Sede nas Copas')
plt.xlabel('País-Sede')
plt.ylabel('Número de Vitórias')
plt.tight_layout()
plt.show()
```

### Existe relação entre país-sede e desempenho do time da casa?

Sim, os dados mostram uma **forte correlação positiva** entre ser o país-sede e ter um desempenho acima da média. Isso pode ser explicado por diversos fatores:

Ao criar um ambiente de apoio, com mais investimentos em infraestrutura e treinamento, o impacto psicológico no desempenho das equipes locais poderia ser significativo, ajudando a promover o esporte em áreas carentes e incentivando o sucesso.

### 3. **Problema Ambiental: Uso de Infraestrutura Sustentável nas Copas**

### 4. **Negócio: Impacto de Tecnologias de Acompanhamento e Análise de Torcida**
   - **Insight**: O apoio da torcida e a influência psicológica que ela exerce sobre a seleção da casa podem ser fatores determinantes.

# Quais seleções são mais constantes no Top 4 (campeão, vice, 3º, 4º)?
- Aponta seleções com maior visibilidade histórica.


```python
from collections import Counter
import matplotlib.pyplot as plt
import pandas as pd

top4_list = []
for _, row in cups.iterrows():
    top4_list.extend([row['Winner'], row['Runners-Up'], row['Third'], row['Fourth']])

top4_count = Counter(top4_list)

df_top4 = pd.DataFrame.from_dict(top4_count, orient='index', columns=['Top 4 Appearances'])
df_top4 = df_top4.sort_values(by='Top 4 Appearances', ascending=False)

ax = df_top4.head(10).plot(kind='bar', figsize=(10,5), colormap='tab10', legend=False)

for p in ax.patches:
    ax.annotate(f'{p.get_height()}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                xytext=(0, 5),
                textcoords='offset points',
                ha='center', va='bottom')

plt.title('Seleções com Mais Presenças no Top 4 da Copa do Mundo')
plt.xlabel('Seleção')
plt.ylabel('Número de Aparições no Top 4')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Quais seleções são mais constantes no Top 4 (campeão, vice, 3º, 4º)?

Sim, os dados mostram que algumas seleções têm **presença recorrente entre as quatro melhores** em diversas edições da Copa, como Alemanha, Brasil, França e Itália. Isso reflete **regularidade de desempenho ao longo do tempo**, o que pode gerar vários insights práticos:

- **Oportunidade de Negócio**: Consultorias esportivas podem usar esses dados para oferecer pacotes de benchmarking a federações menos tradicionais, mostrando práticas bem-sucedidas que contribuíram para a regularidade das grandes seleções.

### 2. **Problema Social: Igualdade de Acesso a Treinamento e Desenvolvimento**
- **Insight**: As seleções mais presentes no Top 4 vêm, em sua maioria, de países com recursos e histórico de investimento esportivo.
- **Oportunidade Social**: Programas internacionais podem ser criados para garantir que seleções de países em desenvolvimento também tenham acesso a tecnologias, centros de treinamento e intercâmbios que ajudem a melhorar sua competitividade a longo prazo.
