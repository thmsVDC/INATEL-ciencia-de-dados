### 1. Por meio do dataset paises.csv, trace dois gráficos de linhas em um mesmo plano cartesiano, um mostrando a taxa de mortalidade (Deathrate) e outro a taxa de natalidade (Birthrate) dos países da América do Norte;

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('paises.csv', sep=';', encoding='utf-8')

df.columns = df.columns.str.replace('\ufeff', '').str.strip()

df['Birthrate'] = pd.to_numeric(df['Birthrate'], errors='coerce')
df['Deathrate'] = pd.to_numeric(df['Deathrate'], errors='coerce')

north_america = df[df['Region'].str.strip() == 'NORTHERN AMERICA']

plt.figure(figsize=(10, 5))
plt.plot(north_america['Country'], north_america['Birthrate'], label='Natalidade', marker='o')
plt.plot(north_america['Country'], north_america['Deathrate'], label='Mortalidade', marker='o')
plt.xticks(rotation=45)
plt.title('Taxa de Natalidade e Mortalidade - América do Norte')
plt.ylabel('Taxa por mil habitantes')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

### 2. Por meio do dataset space.csv, trace um gráfico em barras mostrando quantas empresas espaciais diferentes os EUA e a CHINA possuem;

```python
space = pd.read_csv('space.csv', sep=None, engine='python')

usa = space[space['Location'].str.contains('USA', na=False)]['Company Name'].unique()
china = space[space['Location'].str.contains('China', na=False)]['Company Name'].unique()

plt.bar(['EUA', 'CHINA'], [len(usa), len(china)], color=['blue', 'red'])
plt.title('Nº de Empresas Espaciais - EUA vs China')
plt.ylabel('Quantidade')
plt.show()
```

### 3. Por meio do dataset space.csv, trace um gráfico em torta ilustrando a porcentagem de missões da empresa Roscosmos que deram certo e que deram errado;

```python
roscosmos = space[space['Company Name'] == 'Roscosmos']

sucesso = len(roscosmos[roscosmos['Status Mission'] == 'Success'])
falha = len(roscosmos[roscosmos['Status Mission'] != 'Success'])

plt.pie([sucesso, falha], labels=['Sucesso', 'Falha'], autopct='%1.1f%%', colors=['green', 'gray'])
plt.title('Desempenho das Missões Roscosmos')
plt.show()
```
